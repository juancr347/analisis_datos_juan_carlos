## Numpy dataset bidimensional

Trabajar con dos dimensiones

Leer de URL

```python
import numpy as np
import urllib
```

```python
# https://archive-beta.ics.uci.edu/dataset/53/iris
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
response = urllib.request.urlopen(url)

data = np.genfromtxt(response, delimiter=",", usecols=(0, 1, 2, 3))
data[:10]
```

```python
print(data.ndim)
print(data.shape)
```

```python
# sepal length in cm
print(f"sepal length mean: {np.mean(data[:, 0])}")
print(f"sepal length median: {np.median(data[:, 0])}")
print(f"sepal length max: {np.max(data[:, 0])}")
print(f"sepal length min: {np.min(data[:, 0])}")
print(f"sepal length std: {np.std(data[:, 0])}")
```

```python
# sepal width in cm
print(f"sepal width mean original: {np.mean(data[:, 1])}")
print(f"sepal width mean rounded: {np.mean(data[:, 1]):.2f}")

print(f"sepal width median: {np.median(data[:, 1])}")
print(f"sepal width max: {np.max(data[:, 1])}")
print(f"sepal width min: {np.min(data[:, 1])}")
print(f"sepal width std: {np.std(data[:, 1])}")
```

```python
# petal length in cm
print("petal length statistics:")
print(f"mean rounded: {np.mean(data[:, 2]):.2f}")
print(f"median: {np.median(data[:, 2]):.2f}")
print(f"max: {np.max(data[:, 2]):.2f}")
print(f"min: {np.min(data[:, 2]):.2f}")
print(f"std: {np.std(data[:, 2]):.2f}")
```

## Búsquedas

https://refactoring.guru/refactoring/catalog

```python
# Encontrar el índice de la flor con la longitud del pétalo más larga (índice 2)

PETAL_LENGTH = 2 # position

max_petal_length_index = np.argmax(data[:, PETAL_LENGTH])

max_petal_length = data[max_petal_length_index, PETAL_LENGTH] # data [0, 0]

print("Índice de la longitud del pétalo más larga:", max_petal_length_index)
print("Longitud del pétalo más larga:", max_petal_length)
```

## Ordenar

```python
# ordenar por la primera columna ascendentemente indice 0
# longitudes del sépalo (índice 0)

indices = data[:, 0].argsort() # argsort devuelve índices

data[indices] # filtra por índices
```

```python
np.sort(data, axis=0)  # ordena las columnas
```

```python
np.sort(data[:, 0])
```

```python
# ordenar por la primera columna descendentemente indice 0
# longitudes del sépalo (índice 0)

indices = data[:, 0].argsort() # argsort devuelve índices

data[indices][::-1] # filtra por índices
```
